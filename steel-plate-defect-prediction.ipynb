# 5-数据处理

```python
# 异常值计算公式
def outlier_threshhold(dataframe,column,q1=0.25,q3=0.75):

    Q1=dataframe[column].quantile(q1)

    Q3=dataframe[column].quantile(q3)

    iqr=Q3-Q1

    up_limit=Q3+1.5*iqr

    low_limit=Q1-1.5*iqr

    return low_limit,up_limit

# 异常值计数    
def outlier_percentage(dataframe, column):

    low_limit, up_limit = outlier_threshhold(dataframe, column)

    outliers = [x for x in dataframe[column] if (x > up_limit) or (x < low_limit)]  # 将 | 改为 or

    print(column)  # 将 col 改为 column

    print("Outliers percentage:", len(outliers) / dataframe[column].shape[0] * 100, "%")

    # print("Outliers Count :", len(outliers))

    print("-------------------------")

# 检查异常值 
def check_outliers(dataframe,column):

    low_limit,up_limit=outlier_threshhold(dataframe,column)

    outliers=(dataframe[column]>up_limit) | (dataframe[column]<low_limit)

    if outliers.any():

        return True

    else:

        return False

# 用IQR替换异常值     
def replace_with_threshholds(dataframe,column):

    low_limit,up_limit=outlier_threshhold(dataframe,column)

    dataframe.loc[(dataframe[column]<low_limit),column]=low_limit

    dataframe.loc[(dataframe[column]>up_limit),column]=up_limit

for col in (num_col):

    outlier_percentage(df_train,col)
```

```python
import pandas as pd



def detect_skewness(dataframe, threshold=0.5):

    """

    检测pandas DataFrame中的左偏或右偏列。



    参数:

    dataframe (pandas DataFrame): 要分析的DataFrame。

    threshold (float): 认为列偏斜的阈值。

                      默认值为0.5。



    返回:

    skewed_columns (list): 偏斜列名的列表。

    """

    skewed_columns = []



    for column in dataframe.columns:

        skewness = dataframe[column].skew()

        if abs(skewness) > threshold:

            #skewed.append((column, skewness))

            skewed_columns.append(column)

            

    

    return skewed_columns 

skewed_cols = detect_skewness(df_train[num_col])

print("Skewed columns:", skewed_cols)
```

### 自定义转换器

```python
class OrdinalEncodeColumns(BaseEstimator, TransformerMixin):

    """

    转换器类，用于对Pandas DataFrame的指定列执行序数编码。



    参数

    ----------

    columns : list of str

        要编码的序数列的名称。



    返回

    -------

    pandas.DataFrame

        一个新的DataFrame，其中序数列已编码。

    """

    def __init__(self, columns):

        self.columns = columns

        self.encoder = None

    

    def fit(self, X, y=None):

        ordinal_data = X[self.columns].values

        self.encoder = OrdinalEncoder()

        self.encoder.fit(ordinal_data)

        return self

    

    def transform(self, X):

        X_new = X.copy()

        ordinal_data = X_new[self.columns].values

        encoded_data = self.encoder.transform(ordinal_data)

        X_new[self.columns] = encoded_data

        return X_new

    

    def fit_transform(self, X, y=None):

        self.fit(X)

        return self.transform(X)
```

```python
class LogTransform(BaseEstimator, TransformerMixin):

    """

    对Pandas DataFrame中指定列应用对数变换的转换器类。



    参数

    ----------

    columns : str

        要应用对数变换的列的名称。

    domain_shift : float

        在应用对数变换之前要添加到列中的值。

        

    返回

    ------

        变换后的特征

    """

    def __init__(self, columns, domain_shift=1):

        self.columns = columns

        self.domain_shift = domain_shift



    def fit(self, X, y=None):

        return self



    def transform(self, X):

        X_copy = X.copy()

        X_copy[self.columns] = np.log(X_copy[self.columns] + self.domain_shift)

        return X_copy



    def fit_transform(self, X, y=None):

        return self.transform(X)
```

```python
class StandardScaleTransform(BaseEstimator, TransformerMixin):

    """

    对Pandas DataFrame中指定列应用标准缩放的转换器类。



    参数

    ----------

    cols : list of str

        要应用标准缩放的列的名称。

    """

    def __init__(self, cols):

        self.cols = cols

        self.scaler_ = None



    def fit(self, X, y=None):

        self.scaler_ = StandardScaler().fit(X.loc[:, self.cols])

        return self



    def transform(self, X):

        X_copy = X.copy()

        X_copy.loc[:, self.cols] = self.scaler_.transform(X_copy.loc[:, self.cols])

        return X_copy



    def fit_transform(self, X, y=None):

        self.scaler_ = StandardScaler().fit(X.loc[:, self.cols])

        return self.transform(X)
```

# 7-模型构建

```python
print(f"训练集大小: {X_train_split.shape}")
print(f"测试集大小: {X_test_split.shape}")

# === 彻底清理所有可能冲突的变量 ===
variables_to_clean = ['model', 'feature_importance_model', 'model_name', 'accuracy_scores', 'models']
for var_name in variables_to_clean:
    try:
        if var_name in locals():
            del locals()[var_name]
        if var_name in globals():
            del globals()[var_name]
    except:
        pass

print("已清理所有可能冲突的变量")

# === 重新定义干净的模型列表 ===
models = [
    # 树模型
    DecisionTreeClassifier(random_state=42),  # 决策树分类器
    RandomForestClassifier(random_state=42),  # 随机森林分类器
    GradientBoostingClassifier(random_state=42),  # 梯度提升分类器
    # 线性模型
    LogisticRegression(random_state=42),  # 逻辑回归分类器
    # XGBoost
    XGBClassifier(random_state=42),  # XGBoost分类器
    # CatBoost
    CatBoostClassifier(random_state=42, verbose=False),  # CatBoost分类器
    # 深度学习模型
    MLPClassifier(hidden_layer_sizes=(128, 64, 32), activation='relu', solver='adam', 
                  random_state=42, max_iter=1000, alpha=0.01, early_stopping=True,
                  validation_fraction=0.1, n_iter_no_change=20)  # 多层感知机神经网络
]

# === 验证模型列表 ===
print(f"\n=== 准备评估 {len(models)} 个模型 ===")
model_classes = [model.__class__.__name__ for model in models]
for i, model_class in enumerate(model_classes, 1):
    print(f"{i}. {model_class}")



# === 初始化结果存储变量 ===
from sklearn.metrics import precision_score, recall_score, f1_score

model_name = []
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

print("\n=== 多指标模型评估 (在测试集上) ===\n")

# === 模型评估循环 ===
for idx, model in enumerate(models, 1):
    # 存储当前模型各项指标的列表
    accuracy_scores_model = []
    precision_scores_model = []
    recall_scores_model = []
    f1_scores_model = []
    
    model_name_str = model.__class__.__name__
    print(f"[{idx}/{len(models)}]  正在评估模型: {model_name_str}")
    
    # 对每个缺陷类别进行预测
    for defect_category in ['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults']:
        # 在训练集上训练模型
        model.fit(X_train_split, y_train_split[defect_category])
        
        # 在测试集上进行预测
        y_pred = model.predict(X_test_split)
        
        # 计算各项评估指标
        accuracy = accuracy_score(y_test_split[defect_category], y_pred)
        precision = precision_score(y_test_split[defect_category], y_pred, average='weighted', zero_division=0)
        recall = recall_score(y_test_split[defect_category], y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test_split[defect_category], y_pred, average='weighted', zero_division=0)
        
        # 存储结果（转换为百分比）
        accuracy_scores_model.append(accuracy * 100)
        precision_scores_model.append(precision * 100)
        recall_scores_model.append(recall * 100)
        f1_scores_model.append(f1 * 100)
    
    # 计算各项指标的平均值
    avg_accuracy = np.mean(accuracy_scores_model)
    avg_precision = np.mean(precision_scores_model)
    avg_recall = np.mean(recall_scores_model)
    avg_f1 = np.mean(f1_scores_model)
    
    # 存储结果
    model_name.append(model_name_str)
    accuracy_scores.append(avg_accuracy)
    precision_scores.append(avg_precision)
    recall_scores.append(avg_recall)
    f1_scores.append(avg_f1)
    
    # 打印详细结果
    print(f"   评估结果:")
    print(f"     准确率 (Accuracy):  {avg_accuracy:.2f}%")
    print(f"     精确率 (Precision): {avg_precision:.2f}%")
    print(f"     召回率 (Recall):    {avg_recall:.2f}%")
    print(f"     F1分数 (F1-Score):  {avg_f1:.2f}%")
    print("-" * 60)

# === 验证最终结果 ===
print(f"\n===  多指标评估完成，共评估了 {len(model_name)} 个模型 ===")
print("模型名称列表:", model_name)
print("\n 各项指标概览:")
print(f"准确率范围: {min(accuracy_scores):.2f}% - {max(accuracy_scores):.2f}%")
print(f"精确率范围: {min(precision_scores):.2f}% - {max(precision_scores):.2f}%")
print(f"召回率范围: {min(recall_scores):.2f}% - {max(recall_scores):.2f}%")
print(f"F1分数范围: {min(f1_scores):.2f}% - {max(f1_scores):.2f}%")
```

```python
# 显示详细的多指标模型比较
print("\n===  模型综合评估排行榜 ===")
print("排名".ljust(5) + "模型名称".ljust(25) + "准确率".ljust(10) + "精确率".ljust(10) + "召回率".ljust(10) + "F1分数")
print("=" * 80)

# 按准确率排序显示
sorted_indices = np.argsort(accuracy_scores)[::-1]  # 降序排列

for i, idx in enumerate(sorted_indices):
    rank = "1" if i == 0 else "2" if i == 1 else "3" if i == 2 else f"{i+1}."
    print(f"{rank}".ljust(5) + 
          f"{model_name[idx]}".ljust(25) + 
          f"{accuracy_scores[idx]:.2f}%".ljust(10) + 
          f"{precision_scores[idx]:.2f}%".ljust(10) + 
          f"{recall_scores[idx]:.2f}%".ljust(10) + 
          f"{f1_scores[idx]:.2f}%")

# 找出各项指标的最佳模型
best_accuracy_idx = np.argmax(accuracy_scores)
best_precision_idx = np.argmax(precision_scores)
best_recall_idx = np.argmax(recall_scores)
best_f1_idx = np.argmax(f1_scores)

print(f"\n 各项指标最佳模型:")
print(f" 最佳准确率: {model_name[best_accuracy_idx]} ({accuracy_scores[best_accuracy_idx]:.2f}%)")
print(f" 最佳精确率: {model_name[best_precision_idx]} ({precision_scores[best_precision_idx]:.2f}%)")
print(f" 最佳召回率: {model_name[best_recall_idx]} ({recall_scores[best_recall_idx]:.2f}%)")
print(f" 最佳F1分数: {model_name[best_f1_idx]} ({f1_scores[best_f1_idx]:.2f}%)")

# 综合评分：各项指标的平均值
composite_scores = []
for i in range(len(model_name)):
    composite_score = np.mean([accuracy_scores[i], precision_scores[i], recall_scores[i], f1_scores[i]])
    composite_scores.append(composite_score)

best_composite_idx = np.argmax(composite_scores)
print(f"\n 综合评分最佳模型: {model_name[best_composite_idx]} (综合得分: {composite_scores[best_composite_idx]:.2f}%)")
```
